Today, we'll be going over the basics of unsupervised learning. Recall that in unsupervised learning,  we only provide the model with the input, and leave it to draw its own conclusions for the output. Usually in this case we do not know what the "correct" output looks like, so we would like the model to determine that.

Unsupervised data is all about discovering patterns! In terms of methods, we will mainly focus on clustering, but also speak a bit on dimension reduction.


### Clustering

CLustering is an excellent way to indentify groups in a dataset that aren't known.

**K-Means Clustering**: Finding a specific number of clusters in a sample, and sorting data based on the mean/centroid of each cluster. Therefore, data will be sorted into the group with the closest centroid.

```python
#This is a very simple clustering model to group fruits based on size and sweetness

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
fruits = np.array([
    [5, 8],
    [6, 9],
    [5, 7],
    [9, 2],
    [8, 3],
    [10, 2],
    [1, 6],
    [2, 5],
    [1, 7],
])



model = KMeans(n_clusters=3) #This means that we want data to go into three groups
model.fit(fruits)
```

```python
labels = model.predict(fruits)
print("Cluster labels for le fruits:")
print(labels)
```

The cool this is that once you create the clusters, you can assign new data and have it sorted into the pre-existing clusters!

```python
more_fruits = np.array([
    [6, 8],
    [9, 2],
    [2, 6],
])
new_labels = model.predict(more_fruits)
print(new_labels)
```

I suggest using a scatter plot to visualize the clusters, if your model is good you should be able to pretty much draw circles around the plot!

```python
plt.figure(figsize=(8, 6))
plt.scatter(fruits[:, 0], fruits[:, 1], c=labels, cmap='viridis', s=100, label="Original Fruits")
plt.scatter(more_fruits[:, 0], more_fruits[:, 1], c=new_labels, cmap='viridis', s=200, marker='X', edgecolor='black', label="New Fruits")
plt.title("Fruit Clustering by Size and Sweetness")
plt.xlabel("Size")
plt.ylabel("Sweetness")
plt.legend()
plt.grid(True)
plt.show()
```

Other types of clustering include:

### Heirachical Clustering
Heirarchical clustering involves groups thaat are in other groups, which may then be in other groups, and so on. It can be good to understand how different grroups relate to each other (oft as the name suggests, in some sort of heirarchy).

### DBSCAN Clustering
DBSCAN Clustering is density -based clustering. DBSCAN does not require the specification of k, or the number of groups. Works well for uneven clusters, and on dirty data.

### Dimentionality Reduction
Reducing the dimesionality of data is done by combining variables, and using them to make new ones. There are several methods of doing this, for example:
*  PCA
*  t-SNE
*  MAP
