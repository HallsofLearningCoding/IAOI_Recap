##Classification Models
Classification models are used to assign items to categories, ie, to classify them. Instead of outputing continuous values, classification models output Boolean values: for instance true (1) or false (0). Let's look at how to create a few classification models.

##Logistic Regression
Despite it's name this is a type of classification model. It actually works similar to linear regression, in that instead of fitting a line, the model aims to fit an s-curve.

The result of a logistic regression will be a range between 0 and 1; the closer the number is to 1, the higher the probability that the prediction is true.

Note that this setup would only work to predict two categories (like cat vs dog). If you would like to use this method to predict more than two categories, you can use multinomial logistic regression instead.

Remember when we looked at the titanic dataset? Let's use it to create a model to predict whether or not the students in this class would have survived the titanic!

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import seaborn as sns
df = sns.load_dataset("titanic")
```

```python
df.head()

#Remember, before we do anything we need to:
#1)Explore the data
#2)Prepare the data

#We know we want to predict survivors, so why don't you do a boxplot comparing each feature with survived?


#Next, clean the data
#For example, you may want to remove any na values
print(df.isna().sum())
titanicModel = df.dropna()

print("\n\n\nThis df is a little cleaner")
titanicModel.head()
print(titanicModel.isna().sum())

print("\n\n\nIs there a better way to clean the data? Removing all na values might remove quite a bit!")

#You may also want to convert string to numerical values. Here's an easy way to do that
#This is called encoding
titanicModel = pd.get_dummies(titanicModel, columns=['sex'], drop_first=True)

titanicModel.head()
```

```python
#Let's set the features that we want to feed the model, as well as the target prediction

features = ['pclass', 'age', 'sibsp', 'parch', 'fare', 'sex_male']
target = 'survived'

X = titanicModel[features]
y = titanicModel[target]
```

```python
#Then, like before, we will split the model

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)

```

```python
#Fit the model and check out the predictions!

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(y_pred)
```

Great! We have our output (1 for survived, 0 for...did not survive). How can we better represent the data in a way that makes sense?
